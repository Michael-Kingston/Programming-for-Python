```python
def vis_num_distribution(dataset: pd.DataFrame, column_name: str):
    """
    Visualizes the distribution of a numerical column, split by the "Refund.Request" column.

    Generates a histogram, a split boxplot, and an overall boxplot to provide a comprehensive view
    of the numerical column's distribution. Also prints the min, max, and mean values of the column.

    Args:
        dataset: The Pandas DataFrame containing the data.
        column_name: The name of the numerical column to visualize.

    Returns:
        None. Displays the plots and prints the min, max, and mean values.
    """
    fig, axes = plt.subplots(1, 2, figsize=(15, 6))  # Changed to 1 row and 2 columns

    # Histogram split by Refund.Request with KDE
    sns.histplot(data=dataset, x=column_name, bins=30, ax=axes[0], kde=True, color="#96ffda")
    kde_line = axes[0].lines[0]
    kde_line.set_color("#8c0829")
    kde_line.set_linewidth(2)
    axes[0].set_title(f'Histogram of {column_name}')
    axes[0].set_xlabel(column_name)
    axes[0].set_ylabel('Frequency')

    # Boxplot
    sns.boxplot(x=dataset[column_name], ax=axes[1], color="#96ffda")
    axes[1].set_title(f'Boxplot of {column_name}')
    axes[1].set_xlabel(column_name)

    plt.tight_layout()
    plt.show()
    
    # Print min, max, and mean values
    min_value = dataset[column_name].min()
    max_value = dataset[column_name].max()
    mean_value = dataset[column_name].mean()
    
    print(f"Min value of {column_name}: {min_value}")
    print(f"Max value of {column_name}: {max_value}")
    print(f"Mean value of {column_name}: {mean_value}")
```

```python
vis_num_distribution(costs, "Cost")
```
